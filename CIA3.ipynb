```python
class Customer:
    def __init__(self, custid, custname="Unknown", occup="Unknown", age=None):
        self.custid = custid
        self.custname = custname
        self.age = age
        self.occup = occup
    def display(self):
        print("Customer id ID:", self.custid)
        print("Customer Name:", self.custname)
        print("Customer Age:", self.age)
        print("Occupation:", self.occup)
       
class Policy(Customer):
    allowed_policy=["Health","Life","Vehicle","Property"]
    def __init__(self, custid, pno, ptype, amt, pduration=None):
        super().__init__(custid)
        self.pno = pno
       
        if ptype not in self.allowed_policy:
            raise ValueError(f"Invalid policy. Allowed: {self.allowed_policy}")
        self.ptype = ptype
        self.amt = amt
        self.pduration = pduration
    def display(self):
        print("policy number:", self.pno)
        print("policy type", self.ptype)
        print("premium amt:", self.amt)
        print("policy duration:", self.pduration)
class Agent:
    def __init__(self, ano, aname, loc):
        self.ano = ano
        self.aname = aname
        self.loc = loc
    def display(self):
        print("Agent Id:", self.ano)
        print("Agent Name:", self.aname)
        print("Branch Location:", self.loc)
class InsuranceSystem:
    def __init__(self):
        self.cust = {}
        self.pol = {}
        self.agnt = {}

    def add_customer(self):

        try:
            custid = int(input("Enter Customer ID: "))

            if custid in self.cust:
                raise RuntimeError("Customer ID already exists")

            custname = input("Enter Customer Name: ")
            age = int(input("Enter Customer Age: "))

            if age <= 0:
                raise ValueError("Age must be positive")

            occup = input("Enter Occupation: ")
           

            cust = Customer(custid, custname, age, occup)

            self.cust[custid] = cust

            print("Customer registered successfully")

        except ValueError:
            print("Invalid input. ID and Age must be numeric")

        except RuntimeError as e:
            print(e)

        except Exception as e:
            print("Unexpected error:", e)

    def add_policy(self):
        try:
            custid = int(input("Enter Customer ID for this policy: "))
            if custid not in self.cust:
                print("Error: Customer ID does not exist. Register customer first.")
                return

            pno = int(input("Enter Policy Number: "))
            if pno in self.pol:
                raise RuntimeError("Policy Number already exists")

            ptype = input("Enter Policy Type: ")
            amt = int(input("Enter Amount: "))
            pduration = int(input("Enter Duration : "))

            pol = Policy(custid, pno, ptype, amt, pduration)

            self.pol[pno] = pol
            print("Policy registered successfully")

        except ValueError as e:
            print(f"Input Error: {e}")
        except RuntimeError as e:
            print(e)
        except Exception as e:
            print("Unexpected error:", e)

    def add_agent(self):

        try:
            aid = int(input("Enter agent ID: "))

            if aid in self.agnt:
                raise RuntimeError("agent ID already exists")

            aname = input("Enter Agent Name: ")
            loc = input("Enter loc: ")


            agnt = Agent(aid, aname, loc)

            self.agnt[aid] = agnt

            print("Agent registered successfully")

        except ValueError:
            print("Invalid input. ID must be numeric")

        except RuntimeError as e:
            print(e)

        except Exception as e:
            print("Unexpected error:", e)

    def update_policy(self):

        try:
            pno = int(input("Enter policy no. to update: "))

            if pno not in self.pol:
                raise KeyError("Policy not found")

            ptype = input("Enter ptype: ")
            amt = int(input("Enter amt: "))

            if amt <= 0:
                raise ValueError("amt must be positive")

            pduration = int(input("Enter pduration: "))

            if pduration <= 0:
                raise ValueError("pduration must be positive")

            self.pol[pno].ptype = ptype
            self.pol[pno].amt = amt
            self.pol[pno].pduration = pduration

            print("Policy information updated")

        except KeyError as e:
            print(e)

        except ValueError:
            print("Invalid No.")
           
    def search_policy(self):

        try:
            pno = int(input("Enter Policy no. to search: "))

            if pno not in self.pol:
                raise KeyError("policy not found")

            print("\nPolicy Details")
            self.pol[pno].display()

        except KeyError as e:
            print(e)

        except ValueError:
            print("Invalid No.")

    def display_policies(self):

        if not self.pol:
            print("No Policy records available")
            return

        print("\nAll Policy")

        for pol in self.pol.values():
            print("\n----------------------")
            pol.display()

    def display_agents(self):

        if not self.agnt:
            print("No Agent records available")
            return

        print("\nAll Agent")

        for agnt in self.agnt.values():
            print("\n----------------------")
            agnt.display()
   
    def menu(self):

        while True:

            print("\n===== Insurance Policy Management System =====")

            print("1 Register customer")
            print("2 Add Policy")
            print("3 Update Policy")
            print("4 Search Policy")
            print("5 Display Policy")
            print("6 Add Agent")
            print("7 Display Agent")
            print("8 Exit")

            try:

                choice = int(input("Enter your choice: "))

                if choice == 1:
                    self.add_customer()

                elif choice == 2:
                    self.add_policy()

                elif choice == 3:
                    self.update_policy()

                elif choice == 4:
                    self.search_policy()

                elif choice == 5:
                    self.display_policies()

                elif choice == 6:
                    self.add_agent()

                elif choice == 7:
                    self.display_agents()

                elif choice == 8:
                    print("Exiting system")
                    break

                else:
                    print("Invalid choice")

            except ValueError:
                print("Please enter a valid numeric choice")

system = InsuranceSystem()
system.menu()
```
